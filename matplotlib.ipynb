# matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.show()
```

```python
x = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat']
y = [10,5,50,100,6,9]

plt.title('bakery sales')
plt.xlabel("sale's day")
plt.ylabel('sales')

plt.plot(x, y)
plt.show()
```

# plt.plot(x, y , color="color name", linestyle="line_style", linewidth=value, marker="marker symbol", label="label name")

```python
month = [1, 2, 3, 4]
sale = [1000,500,900,8000]

plt.plot(month, sale, color='red', linestyle='--', linewidth=1, marker='o', label='market sales')
plt.xlabel("months")
plt.ylabel("sales")
plt.title("sales of month")
plt.legend(loc='lower right',fontsize = 10)
plt.grid(color='blue', linestyle=':', linewidth=1)
plt.xlim(1,4)
plt.ylim(0,8000)
plt.xticks([1, 2, 3, 4], ["M1", "M2", "M3", "M4"])
plt.yticks([1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000], ["N1", "N2", "N3", "N4", "N5", "N6", "N7", "N8"])
plt.show()
```

# plt.bar(x, height, color="color name", width, label="label_name")

```python
x = ['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry']
y = [23, 17, 35, 29, 12]

plt.bar(x, y, color="green", label='sales of 2025')
plt.xlabel("fruits")
plt.ylabel('sales')
plt.title('sales of fruits')
plt.legend()
plt.show()
```

# plt.pie(values, labels = label_list, colors = color_list, autopct='%1.1f%%')

```python
regions = ['North', 'West', 'East', 'South']
values = [2500, 3000, 220, 2565] 

plt.pie(values, labels=regions, colors=['red','blue','orange','yellow'], autopct='%1.1f%%')
plt.title("revenue by regions")
plt.show()
```

# scatter plot : scatter plot is graph where each point represent one observation with two values x and y
# plt.scatter(x, y, color="color_name", marker="marker style", label="label name")

```python
hours = [1, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7, 7.5, 8, 9, 10]
marks = [30, 35, 40, 42, 45, 50, 52, 55, 60, 63, 66, 70, 74, 78, 85, 90]

plt.scatter(hours, marks, color='red', marker='^', label='study hour v/s marks')
plt.xlim(1, 11)
plt.ylim(30, 100)
plt.xlabel('hours studied')
plt.ylabel('marks achieved')
plt.title('student study hour v/s marks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
hours_a = [1, 2, 3, 4, 5, 6, 7]
marks_a = [35, 40, 45, 50, 55, 60, 65]

hours_b = [1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5]
marks_b = [30, 38, 44, 53, 58, 66, 72]

plt.scatter(hours_a, marks_a, color='blue', label='Class A')
plt.scatter(hours_b, marks_b, color='orange', label='Class B')
plt.xlim(0, 11)
plt.ylim(25, 100)
plt.xlabel('hours studied')
plt.ylabel('marks achieved')
plt.title('comparision between class A and class B')
plt.legend()
plt.grid(True)
plt.show()
```

# plt.subplot(nrows, ncolumns, index(starts from 1))


```python
a = [1, 2, 3, 4, 5, 6, 7]
b = [38, 30, 55, 40, 85, 10, 35]

plt.subplot(1,2,1) # 1st row, 2nd column, 1st index
plt.plot(a, b)
plt.title("line chart")

plt.subplot(1,2,2) # 1st row, 2nd column, 2nd index
plt.bar(a, b)
plt.title("bar chart")

# plt.tight_layout()
plt.show()
```

# another way to show above
# fig, ax=plt.subplots(nrows, ncolumns, figsize=(width, height))


```python
a = [1, 2, 3, 4, 5, 6, 7]
b = [38, 30, 55, 40, 85, 10, 35]

fig, ax = plt.subplots(1, 2, figsize = (10,5))

ax[0].plot(a, b, color="blue")
ax[0].set_title("line chart")

ax[1].bar(a, b, color="green")
ax[1].set_title("bar chart")

fig.suptitle('comparison between two graphs')

plt.tight_layout()
plt.show()
```
