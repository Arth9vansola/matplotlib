# matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.show()
```

```python
x = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat']
y = [10,5,50,100,6,9]

plt.title('bakery sales')
plt.xlabel("sale's day")
plt.ylabel('sales')

plt.plot(x, y)
plt.show()
```

```python
month = [1, 2, 3, 4]
sale = [1000,500,900,8000]

plt.plot(month, sale, color='red', linestyle='--', linewidth=1, marker='o', label='market sales')
plt.xlabel("months")
plt.ylabel("sales")
plt.title("sales of month")
plt.legend(loc='lower right',fontsize = 10)
plt.grid(color='blue', linestyle=':', linewidth=1)
plt.xlim(1,4)
plt.ylim(0,8000)
plt.xticks([1, 2, 3, 4], ["M1", "M2", "M3", "M4"])
plt.yticks([1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000], ["N1", "N2", "N3", "N4", "N5", "N6", "N7", "N8"])
plt.show()
```
